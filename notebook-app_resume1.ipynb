````python
import requests
import ast

url = "http://127.0.0.1:8000/skill-generator"

payload = {
    "requirements": "Experience with Python, FastAPI, Docker, cloud deployment, and machine learning.",
    "resume_skills": resume_skills
}

try:
    response = requests.post(url, json=payload)
    response.raise_for_status()
    print("Response JSON:", response.json())

    s = response.json()['response']

    # --- Clean up the Markdown formatting ---
    cleaned = (
        s.replace("```python", "")
         .replace("```", "")
         .strip()
    )

    # Convert to Python list
    result = ast.literal_eval(cleaned)

    print("Final list:", result)

except requests.exceptions.RequestException as e:
    print("Request failed:", e)
except Exception as e:
    print("Parsing failed:", e)
````

````python
import json
import ast
import re

failed_list = []
for i in range(df.shape[0]):
    payload = {
        "requirements": df.iloc[i, 4],
        "resume_skills": resume_skills
    }
    try:
        response = requests.post(url, json=payload)
        response.raise_for_status()  # Raises an error for HTTP codes >= 400
#         print("Response JSON:", response.json())
        s = response.json()['response']
        
        # More robust cleaning approach
        # Remove code block markers and language tags
        cleaned = re.sub(r'^```(?:python)?\s*\n?', '', s.strip())  # Remove opening ```python
        cleaned = re.sub(r'\n?```\s*$', '', cleaned)  # Remove closing ```
        cleaned = cleaned.strip()
        
#         print(f"Cleaned string: {repr(cleaned)}")  # Debug print
        
        # Convert the string representation to a real Python list
        result = ast.literal_eval(cleaned)
        
        df.iloc[i, 11] = json.dumps(result)
        print(json.dumps({"row": i, "Skills List":result, "input": df.iloc[i,4]},indent=4))
#         print(f"Row {i} processed successfully")
        
    except requests.exceptions.RequestException as e:
        print(f"row: {i}, Request failed: {e}\n")
        failed_list.append(i)
    except (ValueError, SyntaxError) as e:
        print(f"row: {i}, Parsing failed: {e}, cleaned string: {repr(cleaned)}\n")
        failed_list.append(i)

print(f"Failed rows: {failed_list}")
````
